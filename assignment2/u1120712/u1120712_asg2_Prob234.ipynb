### Approach: We will name states to remember the essence of the history of inputs so far. If you do so, and also keep the sense of where the machine is (in an I state, an IF state or an F state), then you'll have the solution fall out automatically!

Please read the Jove_User_Manual and the advice given there regarding how to write your automaton markdowns, how to place comments, how to group the lines of a DFA per state, and how to order which transition you may first consider.


```python
Even010 = md2mc('''
!!! A DFA that accepts an even number of non-overlapping occurence of 010s.
!!! Zero occurences is accepted.
DFA 

!! Describes an empty string or a string that already contains an even number of
!! occurances of 010
IF    : 0 -> F0
IF    : 1 -> IF     !! we can ignore leading zeros

!! Describes the start of a odd occurance of a sequence of 010
F0    : 0 -> F0 
F0    : 1 -> F01

!! Describes the next step in the start of a odd occurance of 010
F01   : 0 -> S010*  !! the sequence we we are looking for
F01   : 1 -> IF

!! Describes completing a odd occurance followed by zero or more symbols
!! that do not form a sequence of 010
S010* : 0 -> S*0
S010* : 1 -> S010*

!! Describes the start of a even occurance of a sequence lead by zero or more
!! symbols
S*0   : 0 -> S*0
S*0   : 1 -> S*01

!! Describes the next step of a even occurance of a sequence lead by zero or more
!! symbols
S*01  : 0 -> F*010
S*01  : 1 -> S010*

!! Describes the completion of even sequence lead by zero or more syumbols
F*010 : 0 -> F0
F*010 : 1 -> IF

''')
```

```python
dotObj_dfa(Even010)
```

3. Using Jove's markdown, define a DFA for the language of strings over alphabet $\{0,1\}$ that ends in a $010$, calling it Ends010. Then obtain the intersection with the Even010 machine using intersect_dfa.

### Approach: We will define a machine for "ends $010$" as illustrated in SV's part, and negate.

```python
Ends010 =  md2mc('''
!!! A DFA that accepts strings that end in the sequence 010 of abritraty length
DFA

!! Describes a string that does not yet end in a subsequence of 010
I     : 0 -> S0
I     : 1 -> I

!! Describes a string that ends in 0
S0    : 0 -> S0
S0    : 1 -> S01

!! Describes a string that ends in 01
S01   : 0 -> F010
S01   : 1 -> I

!! Describes a string that ends in 010
F010  : 0 -> S0
F010  : 1 -> S01

''')
```

```python
# Let's see the "Ends010" machine
dotObj_dfa(Ends010, FuseEdges=True)
```

## Obtain the intersection of Even010 and Ends010

```python
Even010_Ends010 = intersect_dfa(Even010, Ends010)
dotObj_dfa(Even010_Ends010)
```

## Minimize the intersection of Even010 and Ends010

```python
min_Even010_Ends010 = min_dfa(Even010_Ends010)
dotObj_dfa(min_Even010_Ends010)
```
